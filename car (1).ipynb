```python
import pandas as pd
import re
```

```python
try:
    machine1 = pd.read_excel('Машина 1.xlsx')
    machine2 = pd.read_excel('Машина 2.xlsx')
except FileNotFoundError:
    print("Файл не найден")
    exit()
```

```python
# Проверка соответствия структуры заголовков
if list(machine1.columns) != list(machine2.columns):
    print("Структура заголовков файлов не совпадает")
    exit()

# Сравнение файлов Машина 1 и Машина 2
changes_delete = []
changes_add = []
changes_rename = []

# Поиск удаленных элементов
for index, row in machine1.iterrows():
    if row.to_list() not in machine2.values.tolist():
        changes_delete.append(f"Элемент {row.to_list()} удален")

# Поиск добавленных элементов
for index, row in machine2.iterrows():
    if row.to_list() not in machine1.values.tolist():
        changes_add.append(f"Элемент {row.to_list()} добавлен")

# Поиск изменений в количестве элементов
for column in machine1.columns:
    if machine1[column].count() != machine2[column].count():
        changes_rename.append(f"Изменилось количество элементов в столбце {column}")

# Вывод всех изменений
for change in changes_delete:
    print(change)
print('\n')
for change in changes_add:
    print(change)
print('\n')
for change in changes_rename:
    print(change)
print('\n')
```
